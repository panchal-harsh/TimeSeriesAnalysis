# Stationary Example

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
from pandas import Series
import numpy as np
from pandas import read_csv
```

```python
series = pd.read_csv('daily-total-female-births.csv',header=0, index_col=0,)
```

```python
series.plot()
plt.title('Stationary Series')
plt.xticks(rotation = 45)
plt.show()
```

```python
X = series.values
```

```python
result = adfuller(X)
```

```python
result
```

```python
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
	print('\t%s: %.3f' % (key, value))
```

# Non Stationary Series

```python
series_2 = pd.read_csv('airline-passengers.csv',header=0, index_col=0, squeeze=True)
```

```python
X_2 = series_2.values
```

```python
result_2 = adfuller(X_2)
```

```python
print('ADF Statistic: %f' % result_2[0])
print('p-value: %f' % result_2[1])
print('Critical Values:')
for key, value in result_2[4].items():
	print('\t%s: %.3f' % (key, value))
```

# Differencing

```python
from datetime import datetime
dateparse = lambda x: datetime.strptime('190'+x, '%Y-%m')
series_3 = read_csv('shampoo_sales.csv', parse_dates=[0], header=0, index_col=0, date_parser=dateparse, squeeze=True)
series_3.plot()
```

```python
X_3=series_3.values
```

```python
def difference(dataset, interval=1):
 diff = list()
 for i in range(interval, len(dataset)):
     value = dataset[i] - dataset[i - interval]
     diff.append(value)
 return Series(diff)
```

```python
diff=difference(X_3)
```

```python
plt.plot(diff)
```

```python
diff_2 = series_2.diff()
plt.plot(diff_2)
plt.show()
```

```python
diff_2 = series_2.diff().diff()
plt.plot(diff_2)
plt.show()
```
